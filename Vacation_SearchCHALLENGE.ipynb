```python
# Read file from Part 1
city_vacay_df = pd.read_csv("weather_data/WeatherPy.csv")
city_vacay_df.head()
```

```python
# Ask the customer to add a minimum and maximum temperature value.
min_temp = float(input("What is the minimum temperature you would like for your trip? "))
max_temp = float(input("What is the maximum temperature you would like for your trip? "))
```

```python
# Find out if the vacationer is okay with rain
rain = input("Do you want it to be raining? (yes/no) ")
```

```python
# Find out if the vacationer is okay with snow
snow = input("Do you want it to be snowing? (yes/no) ")
```

```python
# Match criteria to answers
if rain == "no" and snow == "no":
    answer_filter_df = city_vacay_df.loc[(city_vacay_df["Max Temp"]<= max_temp) &
                                    (city_vacay_df["Max Temp"] >= min_temp) &
                                    (city_vacay_df["Rain/Inches"] == 0) &
                                    (city_vacay_df["Snow"] == 0)]

elif rain == "no" and snow == "yes":
    answer_filter_df = city_vacay_df.loc[(city_vacay_df["Max Temp"]<= max_temp) &
                                    (city_vacay_df["Max Temp"] >= min_temp) &
                                    (city_vacay_df["Rain/Inches"] == 0) &
                                    (city_vacay_df["Snow"] == 0)]

elif rain == "yes" and snow == "no":
    answer_filter_df = city_vacay_df.loc[(city_vacay_df["Max Temp"]<= max_temp) &
                                    (city_vacay_df["Max Temp"] >= min_temp) &
                                    (city_vacay_df["Rain/Inches"] == 0) &
                                    (city_vacay_df["Snow"] == 0)]

else:
    answer_filter_df = city_vacay_df.loc[(city_vacay_df["Max Temp"]<= max_temp) &
                                    (city_vacay_df["Max Temp"] >= min_temp) &
                                    (city_vacay_df["Rain/Inches"] == 0) &
                                    (city_vacay_df["Snow"] == 0)]
```

```python
# Show table with city info matching rain/snow criteria
answer_filter_df.head()
```

```python
# Create dataframe based on desired climates
desired_travel_areas_df = answer_filter_df.dropna()
desired_travel_areas_df
```

```python
# New data frame to add hotel column
desiredhotels_df = desired_travel_areas_df[["City", "Country","Max Temp","Weather Desc","Lat","Lng"]].copy()
desiredhotels_df["Hotel Name"] = ""
desiredhotels_df
```

```python
# Add a heatmap of temperature for the vacation spots.
locations = desiredhotels_df[["Lat", "Lng"]]
max_temp = desiredhotels_df["Max Temp"]

# Configure gmaps to use your Google API key.
gmaps.configure(api_key=g_key)
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
heat_layer = gmaps.heatmap_layer(locations, weights=max_temp, dissipating=False,
             max_intensity=300, point_radius=4)
marker_layer = gmaps.marker_layer(locations)
fig.add_layer(heat_layer)
fig.add_layer(marker_layer)

# Call the figure to plot the data.
fig
```

```python
# Adding an info box template
info_box_template = """
<dl>
<dt>Hotel</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Weather Desc} and {Max Temp} °F</dd>
</dl>
"""

# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in desiredhotels_df.iterrows()]
```

```python
# Add a heatmap of temperature for the vacation spots and a pop-up marker for each city.
locations = desiredhotels_df[["Lat", "Lng"]]
max_temp = desiredhotels_df["Max Temp"]
gmaps.configure(api_key=g_key)
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
heat_layer = gmaps.heatmap_layer(locations, weights=max_temp,dissipating=False,
             max_intensity=300, point_radius=4)
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(heat_layer)
fig.add_layer(marker_layer)

# Call the figure to plot the data.
fig
```

```python
# Create the output file (CSV).
output_data_file = "weather_data/WeatherPy_vacation.csv"

# Export the City_Data into a CSV.
desiredhotels_df.to_csv(output_data_file, index_label="City_ID")
```
